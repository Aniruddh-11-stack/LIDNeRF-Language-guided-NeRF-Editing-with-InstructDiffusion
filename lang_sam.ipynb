```python
from PIL import Image
from lang_sam import LangSAM

model = LangSAM()
# image_pil = Image.open("/home/intel-user1/DL/Khushal/bear/images/frame_00082.jpg").convert("RGB")
image_pil = Image.open("/home/intel-user1/DL/Khushal/bear/images/frame_00033.jpg").convert("RGB")
text_prompt = "bear"
masks, boxes, phrases, logits = model.predict(image_pil, text_prompt)
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import cv2
mask = masks.detach().cpu()
mask_array = mask.numpy()
#img_gray = cv2.cvtColor(mask_array, cv2.COLOR_GRAY2BGR)
plt.imshow(mask_array[0],cmap='gray')
```

```python
m = Image.fromarray(mask_array[0])
m.save("bear/masks/frame_00082_mask.jpg")
```

```python
img_np = np.array(image_pil)
image_masked = img_np.copy()
image_masked[np.logical_not(mask_array[0])] = 0
plt.imshow(image_masked)
```

```python
im = Image.fromarray(image_masked)
im.save("bear/masked_pics/frame_00082_masked.jpg")
```
